```python
# Define the directories
base_directory = '/home/ka537/external_SSD/Working/PERSISTER/'
folders = ['0/extracted', '1/extracted', '2/extracted']  # Add more folder names as needed
output_directory = os.path.join(base_directory, 'extracted')

# Set copy_files to True if you want to copy files, False if you want to move them
copy_files = False

# Create the 'Combined' subdirectory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Regex pattern to extract the timepoint from the filenames
pattern = re.compile(r'xy(\d+)_(.+)_T(\d+).png')

# Automatically determine the number of folders from the folders array
num_folders = len(folders)

# Function to move or copy and rename files
def process_folder(folder_path, timepoint_offset, copy_files):
    for file_name in os.listdir(folder_path):
        match = pattern.match(file_name)
        if match:
            fov = match.group(1)
            marker = match.group(2)
            timepoint = int(match.group(3)) + timepoint_offset
            
            # New filename with adjusted timepoint
            new_filename = f'xy{fov}_{marker}_T{str(timepoint).zfill(4)}.png'
            new_filepath = os.path.join(output_directory, new_filename)

            # Copy or move the file to the combined directory with the new name
            source_filepath = os.path.join(folder_path, file_name)
            if copy_files:
                shutil.copy2(source_filepath, new_filepath)
            else:
                shutil.move(source_filepath, new_filepath)

# Start processing folders
timepoint_offset = 0  # Initial offset for folder 0
for i in range(num_folders):
    folder_path = os.path.join(base_directory, folders[i])
    process_folder(folder_path, timepoint_offset, copy_files)

    # After processing each folder, update the timepoint_offset based on the max timepoint of the last folder processed
    max_timepoint = max(
        int(pattern.match(f).group(3)) for f in os.listdir(output_directory)
    )
    timepoint_offset = max_timepoint + 1  # Increment the offset for the next folder
```
